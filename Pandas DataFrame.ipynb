# Handling missing values

```python
df4 = pd.read_csv('D:\\ML FDP\\certi\\student.csv')
df4.head()
```

```python
#handle null values 
df4 = pd.read_csv('D:\\ML FDP\\certi\\student.csv', na_values = ['Not available','Missing Name value'])
df4.tail(5)
```

```python
#handle null values of specific coloumn
df4 = pd.read_csv('D:\\ML FDP\\certi\\student.csv', na_values = {'Email Address':'Not available','Name of student':'Missing Name value'})
df4.tail(5)
```

```python
#to keep null values unchanged to Nan
df4 = pd.read_csv('D:\\ML FDP\\certi\\student.csv', keep_default_na = False)
df4.tail()
```

```python
#to boost performance use na_filter
df4 = pd.read_csv('D:\\ML FDP\\certi\\student.csv', na_filter = True)
df4
```

```python
#to see how many null values are there in dataframe or series
df4.isnull()
```

```python
#to check coont of missing values in each coloumn
df4.isnull().sum()
```

```python
#to count sum of all missing values
df4.isnull().sum().sum()
```

```python
#notnull() is opposite to isnull()
df4.notnull()
```

```python
df4.notnull().sum()
```

```python
df4.notnull().sum().sum()
```
