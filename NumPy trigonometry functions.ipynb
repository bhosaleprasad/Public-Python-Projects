# sin(), cos(), tan() functions

```python
#will plot sin, cos and tan waves using matplotlip library
import numpy as np
import matplotlib.pyplot as plt
```

```python
list1 = np.arange(0,np.pi*6,0.2)
print(list1)
```

```python
#sin(value)
np.sin(180)
```

```python
x_axis = list1
print(x_axis)
```

```python
y_axis_sin = np.sin(x_axis)
print(y_axis_sin)
```

```python
plt.plot(x_axis, y_axis_sin)
plt.show()
```

```python
y_axis_cos = np.cos(x_axis)
plt.plot(x_axis,y_axis_cos)
plt.show()
```

```python
y_axis_tan = np.tan(x_axis)
plt.plot(x_axis,y_axis_tan)
plt.show()
```
